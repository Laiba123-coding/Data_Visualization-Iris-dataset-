```python
import pandas as pd  # For loading and inspecting datasets (Handles structured data in tables rows and columns)
import seaborn as sns   # High-level plotting library for prettier graphs.
import matplotlib.pyplot as plt#Lower-level plotting control (e.g., labels, titles).
```

```python
# Load Iris dataset from seaborn's built-in datasets
iris = sns.load_dataset('iris')
```

```python
print("Shape of dataset:", iris.shape)
print("\nColumn names:", iris.columns.tolist())
print("\nFirst five rows:")
print(iris.head())
```

```python
print("\nDataset info:")
print(iris.info())
```

```python
print("\nSummary statistics:")
print(iris.describe())
```

```python
# Scatter plot: shows the relationship between two continous variables Sepal Length vs Sepal Width (colored by species)
plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", palette="deep", s=80)
plt.title("Sepal Length vs Sepal Width", fontsize=14)
plt.show()
```

```python
# Histogram: Petal Length distribution per species
plt.figure(figsize=(8, 5))
sns.histplot(data=iris, x="petal_length", hue="species", kde=True, palette="deep")
plt.title("Petal Length Distribution", fontsize=14)
plt.show()
```

```python
# Box plot: Petal Width by Species(to check outliers)
plt.figure(figsize=(8, 5))
sns.boxplot(data=iris, x="species", y="petal_width", palette="Set2", hue="species", legend=False)
plt.title("Petal Width by Species", fontsize=14)
plt.show()
```
